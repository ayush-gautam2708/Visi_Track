```python
from facenet_pytorch import MTCNN, InceptionResnetV1
from PIL import Image
import torch
import os
import warnings
import matplotlib.pyplot as plt

# Suppress all warnings
warnings.filterwarnings("ignore")

# Load face detector and embedding model
mtcnn = MTCNN(image_size=160)
resnet = InceptionResnetV1(pretrained='vggface2').eval()

# Set dataset path
dataset_path = "/content/lfw-deepfunneled/lfw-deepfunneled"

# Function to load and split the dataset
def load_dataset(dataset_path):
    persons = [d for d in os.listdir(dataset_path) if os.path.isdir(os.path.join(dataset_path, d))][:50]  # Limit to 50 persons
    train_images = []
    test_images = []

    for person in persons:
        person_folder = os.path.join(dataset_path, person)
        images = [f for f in os.listdir(person_folder) if f.lower().endswith('.jpg')]
        images.sort()

        if len(images) >= 2:
            train_images.append((person, os.path.join(person_folder, images[0])))
            for img in images[1:]:
                test_images.append((person, os.path.join(person_folder, img)))
        else:
            print(f"Warning: Person {person} has only {len(images)} image(s). Skipping.")

    return train_images, test_images

# Load the dataset
train_images, test_images = load_dataset(dataset_path)

# Process training images to extract embeddings
train_embeddings = []
train_labels = []

for label, img_path in train_images:
    img = Image.open(img_path).resize((160, 160), Image.LANCZOS)
    face = mtcnn(img)
    if face is not None:
        emb = resnet(face.unsqueeze(0))
        train_embeddings.append(emb)
        train_labels.append(label)
    else:
        print(f"Face not detected in training image {img_path}. Skipping.")

# Check if we have any training embeddings
if train_embeddings:
    train_embeddings = torch.cat(train_embeddings, dim=0)
else:
    print("No training images with detected faces. Exiting.")
    exit()

# Plot sample original images (first 5 persons)
num_samples = min(5, len(train_images))  # Plot up to 5 persons
plt.figure(figsize=(num_samples * 5, 10))  # Adjust figure size based on number of samples

for i, (label, img_path) in enumerate(train_images[:num_samples], 1):
    img = Image.open(img_path)
    plt.subplot(2, num_samples, i)
    plt.imshow(img)
    plt.title(f"Train: {label}")
    plt.axis("off")

    # Plot corresponding test image if available
    test_img_path = next((path for person, path in test_images if person == label), None)
    if test_img_path:
        img_test = Image.open(test_img_path)
        plt.subplot(2, num_samples, i + num_samples)
        plt.imshow(img_test)
        plt.title(f"Test: {label}")
        plt.axis("off")

plt.tight_layout()
plt.show()

# Process test images and evaluate
correct = 0
total = 0

for label, img_path in test_images:
    img = Image.open(img_path).resize((160, 160), Image.LANCZOS)
    face = mtcnn(img)
    if face is not None:
        emb = resnet(face.unsqueeze(0))
        similarities = torch.nn.functional.cosine_similarity(emb, train_embeddings)
        pred_idx = torch.argmax(similarities).item()
        if train_labels[pred_idx] == label:
            correct += 1
        total += 1
    else:
        print(f"Face not detected in test image {img_path}. Skipping.")

# Compute and print accuracy
if total > 0:
    accuracy = correct / total
    print(f"Accuracy: {accuracy:.4f}")
else:
    print("No test images with detected faces.")
```
